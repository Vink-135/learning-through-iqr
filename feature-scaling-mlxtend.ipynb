# FEATURE SCALING THROUGH MLXTEND:


```python

import pandas as pd 
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
dataset = pd.read_csv(r"/kaggle/input/diabetes-data-set/diabetes.csv")
x = dataset.iloc[:,:1]
y = dataset["Outcome"]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
fs = SequentialFeatureSelector(lr,k_features=1,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_score_
```
